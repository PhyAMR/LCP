```python
def build_model(hp):
    model = XGBClassifier(
        # Definimos los rangos de búsqueda
        n_estimators=hp.Int('n_estimators', 50, 300, step=50),
        max_depth=hp.Int('max_depth', 3, 10),
        learning_rate=hp.Float('learning_rate', 1e-3, 0.3, sampling='log')
    )
    return model

tuner = kt.tuners.Sklearn(
    oracle=kt.oracles.BayesianOptimization(
        objective=kt.Objective('score', 'max'),
        max_trials=10),
    hypermodel=build_model,
    directory='my_dir',
    project_name='xgb_tuning'
)

tuner.search(X_train, y_train)
```
